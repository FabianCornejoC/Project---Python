## Indice

* Count_point:

```python
def count_point(data_frame, region):
    years = range(2016, 2023)
    counts = {}
    if region not in data_frame['Region'].unique():
        raise ValueError("La region no existe")

    for year in years:
        count = data_frame[(data_frame['Region'] == region) &
             (data_frame['Fecha'] >= f'{year}-01-01') &
             (data_frame['Fecha'] <= f'{year}-12-31') &
             (data_frame['Precipitacion'] >= 2)].count()['Precipitacion']
        counts[year]= count
    print(f"Cantidad de dias por año que hubo lluvia mayor a 10mm en \033[im{region}\033[0m:")
    for year, count in counts.items():
        print(f"*{year}: {count}")
```
